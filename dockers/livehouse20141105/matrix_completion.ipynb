```python
import numpy as np
from scipy.io import loadmat
mat = loadmat("image_crop.mat")

im = mat["im"]
m, n, r = im.shape

im_rec = np.zeros([m,n,r])
im_miss = np.zeros([m,n,r])

roi = np.random.rand(m,n) > 0.95

for ii in range(3):
    im_miss[:,:,ii] = im[:,:,ii]*roi

for ii in range(3):
    im2 = im_miss[:,:,ii]
    im3 = np.zeros([m,n])
    lambd = 0.1
    error = 1000.0
    
    while error > 1:
        for jj in range(20):
            temp = roi*im2 + (1-roi)*im3;
            U, S, V = np.linalg.svd(temp,full_matrices=False)
            im3 = np.dot(np.dot(U,np.diag(np.maximum( S-1.0 / lambd,0))),V)
            
        error = np.linalg.norm((im2-im3)*roi)
        print "error = ",error
            
        lambd = lambd + 0.1*error
        
    im_rec[:,:,ii] = im3
    
    
    
```
